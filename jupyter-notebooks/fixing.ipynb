```python
import pandas as pd
import os
```

```python
folders = [
    "/hdd/results-analysis/data/processed_data/mcs7/results-set/analytical-7mcs-20dbm",
    "/hdd/results-analysis/data/processed_data/mcs7/results-set/analytical-7mcs-23dbm",
    "/hdd/results-analysis/data/processed_data/mcs7/results-set/Simulated-MCS7-20dbm",
    "/hdd/results-analysis/data/processed_data/mcs7/results-set/Simulated-MCS7-23dbm"
]

for folder in folders:
    print(folder)
    overall_error = pd.DataFrame(columns=["Mean", "Distance"])
    for file in os.listdir(folder):
        print(file)
        if "PDR" in file or "overall" in file:
            continue
            
        df = pd.read_csv("{}/{}".format(folder, file))
        if overall_error.empty:
            overall_error["Mean"] = df["Mean"]
            overall_error["Distance"] = df["Distance"]
        else:
            overall_error["Mean"] = overall_error["Mean"] + df["Mean"]
        
    overall_error.to_csv("{}/overall_error.csv".format(folder), index=False)
```
